```python
import pandas as pd
from babel.numbers import format_currency

kodeRUP = "D197"

## Data E-KATALOG
df_kat = pd.read_excel('katalogdetail.xlsx')
df_prod = pd.read_excel('prodkatalog.xlsx')

## Data E-KATALOG Filter
df_kat_loc = df_kat[df_kat['kd_klpd'] == kodeRUP]
df_kat_loc_lokal = df_kat_loc[df_kat_loc['jenis_katalog'] == "Lokal"]
df_prod_loc = df_prod[df_prod['kd_klpd'] == kodeRUP]
```

```python
## Data E-Katalog
jumlah_produk = df_prod_loc['nama_produk'].count()
jumlah_penyedia = df_prod_loc['nama_penyedia'].value_counts().shape

jumlah_trx_lokal = df_kat_loc_lokal['no_paket'].value_counts().shape
nilai_trx_lokal = df_kat_loc_lokal['total_harga'].sum()
```

```python
print(f"Jumlah Produk    : {jumlah_produk}")
print(f"Jumlah Penyedia  : {jumlah_penyedia[0]}")
print(f"Jumlah Transaksi : {jumlah_trx_lokal[0]}")
print(f"Nilai Transaksi  : {format_currency(nilai_trx_lokal, 'Rp. ', locale='id_ID')}")
```
